**任务1.1：** 导入库以及数据

**任务2.2: **逻辑读取（Logical Indexing）

**任务2.3: **分组读取

1. 对 `release_year` 进行分组，使用 [`.agg`](http://pandas.pydata.org/pandas-docs/version/0.22/generated/pandas.core.groupby.DataFrameGroupBy.agg.html) 获得 `revenue` 的均值。
2. 对 `director` 进行分组，使用 [`.agg`](http://pandas.pydata.org/pandas-docs/version/0.22/generated/pandas.core.groupby.DataFrameGroupBy.agg.html) 获得 `popularity` 的均值，从高到低排列。

要求：使用 `Groupby` 命令实现。

**任务3.1：**对 `popularity` 最高的20名电影绘制其 `popularity` 值。

```python
# 获取数据
movie_data_popularity = movie_data[['original_title', 'popularity']].sort_values(by='popularity', ascending=False).head(20)

plt.figure(figsize = [18, 5])

# 直方图
bins = np.arange(movie_data_popularity['popularity'].min()-1, movie_data_popularity['popularity'].max()+1, 1/2)
plt.hist(data = movie_data_popularity, x = 'popularity', bins = bins)
# plt.xscale('log')



```

**任务3.2：**分析电影净利润（票房-成本）随着年份变化的情况，并简单进行分析。

```python
movie_data_profit = movie_data.eval('profit = revenue_adj - budget_adj').groupby(['release_year'])['profit'].sum().reset_index()

plt.figure(figsize = [18, 5])
plt.errorbar(data = movie_data_profit, x = 'release_year', y = 'profit')
plt.xticks(range(movie_data_profit['release_year'].min() - 1, movie_data_profit['release_year'].max() + 2, 2))
plt.xlabel('release year')
```

### 分析：
    整体上看 电影净利润 和 年份 是正相关的，随着年份的增加 电影净利润逐年增加，但是也存在 一些低谷的年份 如：1966，1976，1991，2000，2006 等

**[选做]任务3.3：**选择最多产的10位导演（电影数量最多的），绘制他们排行前3的三部电影的票房情况，并简要进行分析。

```python
prolific_directors = movie_data.groupby(["director"]).size().reset_index(name='Size').sort_values(by='Size', ascending=False).head(10)
movie_data_revenue = movie_data[movie_data['director'].isin(prolific_directors["director"])].sort_values(by=['director','revenue'], ascending=False).groupby(["director"]).head(3)[['director','revenue','original_title']]
movie_data_revenue['original_title'] = [j for i in range(1,11) for j in range(1,4)]
movie_data_revenue
plt.figure(figsize = [18, 5])
sb.barplot(x="director",y="revenue",hue="original_title",data=movie_data_revenue)
```

### 分析：
    多数导演票房最好的电影都有一枝独秀的情况，超过他自己其它电影票房太多，这种现象只有在导演 Wes Craven 上不太明显，Steven Spielberg 导演的电影票房远远超过的其他导演，而 Woody Allen 导演的电影的票房是 所有导演中最少的。

**[选做]任务3.4：**分析1968年~2015年六月电影的数量的变化。

```python
movie_data_change = movie_data[['original_title', 'release_date', 'release_year', 'genres']]
movie_data_change['release_date'] = pd.to_datetime(movie_data_change['release_date'].str.slice(0,-2) + movie_data_change['release_year'].map(str))

movie_data_change = movie_data_change[(movie_data_change["release_date"] >= pd.Timestamp('1968-01')) & (movie_data_change["release_date"] <= pd.Timestamp('2015-06'))]

movie_data_change['release_month'] = movie_data_change["release_year"].map(str) + "-" + movie_data_change['release_date'].dt.month.map(str)
month = movie_data_change.groupby(by='release_month').size()
month = pd.DataFrame({'month':month.index,'numbers':month.values})
plt.figure(figsize = [18, 5])
plt.errorbar(data = month, x = 'month', y = 'numbers')
```
